```python
import pandas as pd
df = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vTGU1m6KdC2laAwfZZRLkwyEJpzk1uwqf1BNACja06uKpx8L-YHrPzsRF7iQepzTHcBM9Qt2VJlKcsb/pub?output=csv")
```

CREATE AGE BINS/ CATEGORIES

```python
age_bins = [0, 17, 25, 30, 35, 100]
age_labels = ["<18", "18–25", "26–30", "31–35", ">35"]

df["age_group"] = pd.cut(df["Age"], bins=age_bins, labels=age_labels)
```

DETERMINE PERCENT OF PARTICIPANTS IN A GIVEN VARIABLE CATEGORY

```python
def percentage_table(series):
    return (
        series.value_counts(normalize=True)
        .mul(100)
        .reset_index()
        .rename(columns={"index": "Category", series.name: "Category"})
    )
```

CREATE VARIABLE/CATEGORY/PERCENT TABLES FOR EACH GROUP

```python
gender_table = percentage_table(df["Gender"])
age_table = percentage_table(df["age_group"])
status_table = percentage_table(df["Subjective Social Status"])
status_table["Category"] = sorted(status_table["Category"])
status_table
```

LABEL TABLE

```python
def add_variable_name(df, variable):
    df.insert(0, "Variable", variable)
    df.loc[df.index[1:], "Variable"] = ""
    return df

gender_table = add_variable_name(gender_table, "Gender")
age_table = add_variable_name(age_table, "Age")
status_table = add_variable_name(status_table, "Perceived Social Status")
```

CONCATENATE TABLE

```python
final_table = pd.concat(
    [gender_table, age_table, status_table],
    ignore_index=True
)

final_table.rename(columns={"Variable": "Variable", "Category": "Category", "proportion": "Percentage (%)"})
```
